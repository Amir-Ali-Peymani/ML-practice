```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
columns = ["flength", "fwidth", "fsize", "fconc", "fconc1", "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"]
df = pd.read_csv("magic04.data", names=columns)
df.head()
```

```
df["class"] = (df["class"] == "g").astype(int)
```

```
df.head()
```

```
for lable in columns:
  plt.hist(df[df["class"]==1][lable], color ="blue", label="gamma" , alpha=0.7, density=True)
  plt.hist(df[df["class"]==0][lable], color ="red" , label="hadron", alpha=0.7, density=True)
  plt.title(lable)
  plt.ylabel("probility")
  plt.xlabel(lable)
  plt.legend()
  plt.show()
```
