```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,1,2])
plt.ylabel("y numbers")
plt.xlabel("x numbers")
```

```python
numbers = np.arange(0, 5, 0.5)
plt.plot(numbers, numbers**numbers-11, "r--", numbers, numbers**3, "b--")
```
